```python
rfm=pd.merge(orders,order_items, on="order_id")
```

```python
rfm["order_date"]=pd.to_datetime(rfm["order_date"])
```

```python
rfm["total_amount"]=(rfm["list_price"] *
rfm["quantity"])-rfm["discount"]                     
```

```python
today=rfm["order_date"].max()+ pd.Timedelta(days=1)
```

```python
rfm_data = rfm.groupby("customer_id").agg({
    "order_date": lambda x: (today - x.max()).days,
    "order_id": "nunique",
    "total_amount": "sum"
}).reset_index()                                       
```

```python
rfm_data.columns = ["customer_id", "Recency", "Frequency", "Monetary"]
```

```python
from sklearn.cluster import KMeans
```

```python
X = rfm_data[["Recency","Frequency","Monetary"]]
```

```python
kmeans = KMeans(n_clusters=3,
random_state=42)
```

```python
rfm_data["customer_segment"] = Kmeans.fit_predict(X)
```

```python
rfm_data.head()
```

```python
rfm_data.to_sql("customer_segment",
con=engine,if_exists="replace",
index=False)
```
